```python
import os 
import pandas as pd
```

```python
#Read our file
df = pd.read_parquet(r"C:\Users\Quantam Waffle\OneDrive\Desktop\Data engineering project\team_1.parquet") 
print(df.head())
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
#Converting into Datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
#Extracting year, month and day
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
```

```python
base_path = r"C:\Users\Quantam Waffle\OneDrive\Desktop\Data engineering project"

grouped = df.groupby(['year','month','day'])

for (year,month,day), group in grouped:
    folder_path = f'{base_path}/{year}/{month:02d}/{day:02d}'
    os.makedirs(folder_path,exist_ok=True)

    file_path = f'{folder_path}/date.cs'

    group.to_csv(file_path, index = False)

print('Done partitioning data')    
```
